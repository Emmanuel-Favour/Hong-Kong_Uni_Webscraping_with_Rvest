##WEB SCRAPING PROJECT WITH R (rvest).##

The site being scraped is the official site of the University of Hong-kong.

Case-Study: The CS department requires us to obtain department information for further analysis.
(a) Submitting an R program to find all professors (only academic staff) and their corre- sponding personal pages in the link
(link removed).
Present your results with the format “url:name” per line.

```r
#Installing the rvest package
install.packages("rvest")
#Loading the rvest library......
library(rvest)


```

```r

url <- "https://www.cs.hku.hk/people/academic-staff"
#Reading the provided URL
html_content <- read_html(url)

# Extract positions from h4 tags
positions <- html_content %>%
  html_nodes("div.profile_view h4:contains('Professor')") %>%
  html_text()

# Extract professor names from strong tags within div with class 'profile_view'
professor_names <- html_content %>%
  html_nodes("div.profile_view h4 a strong") %>%
  html_text()

# Extract URLs for the professors
professor_urls <- html_content %>%
  html_nodes("div.profile_view h4 a[href*='/index.php/people/']") %>%
  html_attr("href") %>%
  paste0("https://www.cs.hku.hk", .)

# Check if positions exist
if (length(positions) > 0) {
  # Print position, professor name, and professor URL
  for (i in seq_along(positions)) {
    cat("Position:", positions[i], "\n")
    cat("Professor Name:", professor_names[i], "\n")
    cat("Professor URL:", professor_urls[i], "\n\n")
  }
} else {
  cat("No professor positions found.\n")
}

length(positions)
```

```r
library(rvest)
library(purrr)
url <- "https://www.cs.hku.hk/people/academic-staff"
html_content <- read_html(url)

# Extract positions from h4 tags
positions <- html_content %>%
  html_nodes("div.profile_view h4:contains('Professor')") %>%
  html_text()

# Extract professor names from strong tags within div with class 'profile_view'
professor_names <- html_content %>%
  html_nodes("div.profile_view h4 a strong") %>%
  html_text()

# Extract URLs for the professors
professor_urls <- html_content %>%
  html_nodes("div.profile_view h4 a") %>%
  html_attr("href") %>%
  keep(~grepl("/index.php/people/", .)) %>%
  paste0("https://www.cs.hku.hk", .)

# Check if positions exist
if (length(positions) > 0) {
  # Print position, professor name, and professor URL
  for (i in seq_along(positions)) {
    cat("Position:", positions[i], "\n")
    cat("Professor Name:", professor_names[i], "\n")
    cat("Professor URL:", professor_urls[i], "\n\n")
  }
} else {
  cat("No professor positions found.\n")
}

max_length <- max(length(positions), length(professor_names), length(professor_urls))

professors_df <- data.frame(
  Position = c(positions, rep(NA, max_length - length(positions))),
  Professor_Name = c(professor_names, rep(NA, max_length - length(professor_names))),
  Professor_URL = c(professor_urls, rep(NA, max_length - length(professor_urls)))
)

# Save the data frame to a CSV file
write.csv(professors_df, file = "professors_data.csv", row.names = FALSE)
```
